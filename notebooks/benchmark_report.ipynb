# Apex RAG Benchmark
Naive RAG vs Apex RAG comparison.

```python
import json, pandas as pd, matplotlib.pyplot as plt
data = json.load(open('data/eval_runs/benchmark_1779428153.json'))
df = pd.DataFrame({v: data['runs'][v]['metrics'] for v in data['runs']}).T
df.plot(kind='bar', figsize=(10,5), title='RAGAS metrics: naive vs apex')
plt.tight_layout(); plt.savefig('benchmark_metrics.png'); plt.show()
```
